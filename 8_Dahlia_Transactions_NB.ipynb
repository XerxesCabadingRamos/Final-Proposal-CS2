```python
ref = db.reference("transactions")

while True:
    print("\n===== TRANSACTION DATABASE MENU =====")
    print("1. Display Transactions")
    print("2. Add Transaction")
    print("3. Update Transaction")
    print("4. Delete Transaction")
    print("5. Exit")

    choice = input("Enter choice: ")

    # 1. DISPLAY
    if choice == "1":
        transactions = ref.get()
        if transactions:
            print("\n--- Current Transactions ---")
            for tid, data in transactions.items():
                customer = data.get('customer_name', 'N/A')
                item = data.get('MAIN', {}).get('item', 'N/A')
                status = data.get('MAIN', {}).get('status', 'N/A')
                print(f"ID: {tid} | Customer: {customer} | Item: {item} | Status: {status}")
        else:
            print("No transactions found.")

    # 2. ADD
    elif choice == "2":
        tid = input("Enter Transaction ID (e.g., 1007): ")
        name = input("Enter Customer Name: ")
        item = input("Enter Item (e.g., Iced Latte): ")
        price = int(input("Enter Price: "))
        qty = int(input("Enter Quantity: "))
        location = input("Enter Location: ")

        new_transaction = {
            "transaction_id": int(tid),
            "customer_name": name,
            "MAIN": {
                "item": item,
                "quantity": qty,
                "price": price,
                "status": "Pending"
            },
            "SIDE": {
                "location": location,
                "time": "2026-03-30 08:00 PM",
                "buyer_preferences": input("Enter Preferences: ")
            }
        }

        ref.child(tid).set(new_transaction)
        print(f"Transaction {tid} added successfully!")

    # 3. UPDATE
    elif choice == "3":
        tid = input("Enter ID of transaction to update: ")
        target_ref = ref.child(tid)

        if target_ref.get():
            new_status = input("Enter new status (Completed/Pending/Cancelled): ")
            target_ref.child("MAIN").update({
                "status": new_status
            })
            print("Status updated successfully!")
        else:
            print("Transaction not found.")

    # 4. DELETE
    elif choice == "4":
        tid = input("Enter ID to delete: ")
        if ref.child(tid).get():
            ref.child(tid).delete()
            print(f"Transaction {tid} deleted.")
        else:
            print("Transaction not found.")

    # 5. EXIT
    elif choice == "5":
        print("Closing Transaction Manager...")
        break

    else:
        print("Invalid choice. Please pick 1-5.")
```
