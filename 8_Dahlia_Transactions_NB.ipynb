```python
from firebase_admin import db

def transaction_plus(transaction_id, name, amount, date):
    new_data = {
        "name": name,
        "amount": amount,
        "date": date
    }

    ref = db.reference("transactions")
    ref.child(transaction_id).set(new_data)

    print("Transaction added")
```

```python
def transaction_question(transaction_id):
    ref = db.reference("transactions")
    result = ref.child(transaction_id).get()

    if result:
        print("Transaction Found:")
        print(result)
    else:
        print("No Transaction Found.")
```
